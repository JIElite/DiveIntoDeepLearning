```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

```python
x = torch.arange(-8., 8., .1, requires_grad=True)
y = torch.relu(x)
d2l.plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(5, 2.5))
```

```python
y.backward()
```

```python
# y.backward(torch.ones_like(x)) # Work
y.backward(torch.ones_like(x), retain_graph=True) # Retrain_graph 會保留之前 gradient 的結果，使可以執行多次 backward
d2l.plot(x.detach(), x.grad, 'x', 'grad of relu', figsize=(5, 2.5))
```

```python
# Sigmoid
x_ = torch.arange(-8., 8., 0.1, requires_grad=True)
y_ = torch.sigmoid(x_)
d2l.plot(x_.detach(), y_.detach(), 'x', 'sigmoid(x)', figsize=(5, 2.5))
```
